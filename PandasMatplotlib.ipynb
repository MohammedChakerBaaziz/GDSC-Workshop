### 2. Selecting from a DataFrame

One of the first things we need to learn in working with data is how to get and select the data that is most relevant to our case study or problem. With Pandas, we can do that easily using many ways and methods  

#### **Getting Columns**

For a dataframe, passing one label using the **dot** or **[]** returns a **Series**:

```python
# we assign the age column to 'age' and season column to season
age = shopping_behavior.Age
season = shopping_behavior['Season']
# let's print them out
print("age:\n", age.head())
print("\nseason:\n", season.head())
```

**little task:** let's select the item purchased by the first customer:

```python
# first, we select the target column 'Item Purchased'
purchased_items = shopping_behavior['Item Purchased']
# now, to get the item purchased by the first customer, we select the first purchased item (at index=0)
first_purchased_item = purchased_items[0]
# lastly, print it out
first_purchased_item
```

```python
# alternatively
first_purchased_item = shopping_behavior['Item Purchased'][0]
first_purchased_item
```

#### **Index-based Selection**

Attribute selection (Getting Columns) is nice. However, for more advanced operations, we should use`loc` and `iloc`.  
for **Index-based Selection**, we select data based on their numerical positions (indices) in the DataFrame using `iloc` like this: `data.iloc[rows, columns]`

* Select the first row:

```python
# first row => index=0
first_row = shopping_behavior.iloc[0] # not specified column index => all columns
first_row
```

* Select the Gender of the customer of the third purchased item:

```python
# third purchased item => row index = 2
# gender of the customer is the third column => column index = 2
third_customer_gender = shopping_behavior.iloc[2, 2]
third_customer_gender
```

* Select the first 10 purchased items:

```python
first_ten_items = shopping_behavior.iloc[:10, :] # can remove ':' in columns
first_ten_items
```

**Important Note:** in indexing, Python allow us to index easily using the `:`, here are some examples:
* `:100`: first 100 (100 excluded)
* `100:`: from index 100 to last item
* `1:100`: from index 1 to 99
* `:`: everything
* `::-1`: in reversed order*
* `-20:0`: last 20 items
* `:100:4`: from first item till item at index 99 with a step size of 4

* Select first three columns (Customer ID, Age, Gender) of last ten rows:

```python
# first three columns => column index = :3
# last 10 rows => row index = -10:
last_ten_customers = shopping_behavior.iloc[-10:, :3]
last_ten_customers
```

* Select the row at indices 1, 10, 100, 1000:

```python
# rows indices must take an array of the specific indices
specific_indices = shopping_behavior.iloc[[1, 10, 100, 1000], :]
specific_indices
```

#### **Label-based Selection**

for **Label-based Selection**, we select data based on their labels in the DataFrame using `loc` like this: `data.loc[rows, columns]`

* Select the Gender of the customer of the third purchased item:

```python
# third purchased item => row label is 2 (same as index in this case)
third_customer_gender = shopping_behavior.loc[2, 'Gender']
third_customer_gender
```

* Select the columns {Customer ID, Age, Gender} of first ten rows:

```python
# columns {Customer ID, Age, Gender} => pass them as an array
# first 10 rows => row label => :10 (from first label to label 10 included unlike iloc)
first_ten_customers = shopping_behavior.loc[0:10, ['Customer ID', 'Age', 'Gender']]
first_ten_customers
```

#### **Key Difference between `iloc` and `loc`:**

`iloc`: uses the Python stdlib indexing scheme, first element included and last one is excluded. So `0:10` will select entries from index 0 to index 9.  
`loc`: it indexes inclusively. `0:10` will select entries labeled from 0 to 10.

#### **Conditional Selecting:**

To do even more interesting selecting based on some condition(s), we use **conditional selecting**.  
* For example you want to analyse items purchased by people under a certain age (let us say for people under 20 y/o):

```python
shopping_behavior.Age < 20
```

As we notice, this condition operation produces a series of `True/False` booleans based on the age for each row. To get the rows that only satisfy this condition, we may use `loc`:

```python
shopping_behavior.loc[shopping_behavior.Age < 20]
```

Alternatively, we can directly pass the condition to the DataFrame as follows:

```python
shopping_behavior[shopping_behavior.Age < 20]
```

Pandas comes with even more built-in conditional selectors, one of which is `isin` that takes a list:

* For example, Select the **Jackets, Hoodies and Sweaters** items purchased by people **between the age of 20 and 30 inclusive** into a DataFrame called **tops_items_20s**:

```python
tops_items_20s = shopping_behavior[(shopping_behavior['Item Purchased'].isin(['Jacket', 'Hoodie', 'Sweater'])) & ((shopping_behavior.Age >= 20) & (shopping_behavior.Age <= 30))]
tops_items_20s
```

For better code readibility, it is preferable to split the condition and pass it to the selection line:

```python
# Jackets, Hoodies and Sweaters condition
tops = shopping_behavior['Item Purchased'].isin(['Jacket', 'Hoodie', 'Sweater'])
# Age condition
twenties = (shopping_behavior.Age >= 20) & (shopping_behavior.Age <= 30)
# pass it to the selection
tops_items_20s = shopping_behavior[tops & twenties]
tops_items_20s
```

**Important Note:** we used `&` which stands for the logical **AND**, for the logical **OR**, use `|`

### 3. Grouping and Sorting

#### **Grouping**

We often want to group our data based on some **categorical** column and then do something particular to those groups of data. Pandas offer a very powerful tool `groupby()` to achieve that.  
Let's clearify it by the following examples:

* Find the number of purchased items for each item Category using Groupwise analysis:

```python
# one way to do it:
categories_count = shopping_behavior.Category.value_counts()
categories_count
```

this, indicates that we have 1771 rows with Category="Clothing"...  
Yet, we want to do it more efficiently using `groupby()`:

```python
categories_count = shopping_behavior.groupby('Category')['Category'].count()
categories_count
```

`groupby()` created a group for each unique value in the column **Category**. Then, we grabbed the **Category** column and counted how many times it appeared for each group.  

Not just this, we can use any summery functions (mean, max, min...) alongside `groupby()`:

* Find the average rate for each Type (column: **Item Purchased**) of clothes:

```python
# column 'Review Rating' to find the rates
# use the mean() to get the average
categories_average_rate = shopping_behavior.groupby('Item Purchased')['Review Rating'].mean()
categories_average_rate
```

Pandas allow us to even `groupby()` more than just one column. To do that, we pass the list of columns to the `groupby()` function. Let's see an example, shall we?

* Find the total purchase amount for each clothing item category per each season:

```python
# initialize the list of the categories we want to group by (each category per season)
columns_list = ['Season', 'Category']

# to get the total amount, we sum the column 'Purchase Amount (USD)' of all rows for each group
total_category_season = shopping_behavior.groupby(columns_list)['Purchase Amount (USD)'].sum()

# print
total_category_season
```

#### **Sorting**

You already guessed it right from the name. We can store data objects (rows) using `sort_values()` function.  
By default, it will sort the values in an **ascending** order. In case we need the opposite, we must pass `False` to the `ascending` argument, here are some examples:

* Select five purchased items with the highest purchase amount:

```python
# sort objects in a descending order
most_expensive_items = shopping_behavior.sort_values(by='Purchase Amount (USD)', ascending=False)
# display first five items
most_expensive_items.head()
```

* Select the lowest five rates given by reviewers:

```python
# sort the rates
lowest_rates = shopping_behavior['Review Rating'].sort_values()
lowest_rates.head()
```

**Notes:**
* when `sort_values()` is applied to a specific column, no need to specify `by=...`. Here it is different, we are not selecting whole object (row) like the first question, we are only selecting values (cells).
* we can give the argument `by` a list of clumns, just like **Groupng**.

* Sort the types of items (column: Item Purchase) based on their average rating:
recall that we already used `groupby()` to find the average rating for each item type. Now we only sort them:

```python
categories_average_rate.sort_values(ascending=False)
```

### 4. Missing and Wrong Values

#### **Missing Data**

Missing data are entries given the value `NaN`, short for 'Not a Number'. They occur due to various reasons. From which we state the fact that the data instant is either not captured or never existed from the first place. Additionally, when we collect data from multiple sources and combine them together, it is common to encounter `NaN` values due to inconsistencies and incomplete records (objects).

To detect missing values in a dataset, we use the pandas function `isna()` as follows:

```python
shopping_behavior.isna()
```

The `isna()` returns a boolean for each data cell **False** if it is not `Nan` and **True** otherwise.

The previous line of code did not give us much information. We would want to know what are the columns for which there are missing values. To display that, we **sum** the missing values for each column:

```python
shopping_behavior.isna().sum()
```

Great, now we know that columns like **Age, Category** have no `NaN` values. Unlike **Discount Applied** which has a total of 229 missing values.  
The column **Previous Purchases** catches our attention, as it has a total of 3980 of `NaN` values. which is the same number of rows in our dataset (we will think of a way to fix it later on).

We can add `sum()` to get the total of `NaN` values in our datasets (it sums the number of missing values of all columns)

```python
# the first sums the `NaN`s for each column and the second sums the sum of `NaN`s of each column
total_missing_values = shopping_behavior.isna().sum().sum()
print("total missing values: ", total_missing_values)
```

#### **Dropping Missing Values**

In some cases like not having a clear idea and could not figure out why our data is missing. One option is to drop it. Either drop the objects (rows) or attributes (columns).  
To drop data using Pandas. the function `dropna()` will do the work for us. We only have to make sure to give it an `axis` argument:  
**0** if we want to drop **rows** (which is the default if we did not specify `axis`.  
**1** if we aim to remove columns (not so recommended, as we might lose the whole dataset if each colum has at least one `NaN` value).

* Column: **Previous Purchase**

We already said that all rows are `NaN` for this column. Thus, it is more preferable to drop it all.  
`drop()` allow us to remove columns, we only have to give it the list of columns we wnt to remove:

```python
shopping_behavior = shopping_behavior.drop(columns=['Previous Purchases'])
# print all columns to check
shopping_behavior.columns
```

* Column: **Promo Code Used**

We have only three rows with `NaN` for this column. Thus, we can drop them (though, not preferable for small datasets like the one we are working on. It is just for the sake of the example)

```python
shopping_behavior_objects_dropped = shopping_behavior.dropna(subset=['Promo Code Used'])
shopping_behavior = shopping_behavior_objects_dropped
# print the total `NaN` values for the column
shopping_behavior['Promo Code Used'].isna().sum()
```

As you see, we now have zero missing values for the column.
First, we dropped rows that have a `NaN` value in the column given to subset list (Promo Code Used, we can give more than one column). and then we assigned the new dataset to the previous one (previous = new_after_droping_data)  
**Note:** we had to re assign the dataset because we edited the shape of it.

#### **Filling Missing Values**

Another option is to try to somehow fill the data that is missing. Using some techniques or algorithms we can achieve that with the help of `fillna()` function.

* Column: **Payment Method**

let us display the unique values of the column:

```python
shopping_behavior['Payment Method'].unique()
```

Since we are not sure what is the correct value for the missing Payment methods, let's fill them with "Unknown":

```python
shopping_behavior['Payment Method'] = shopping_behavior['Payment Method'].fillna("Unknown")
# let's make sure
shopping_behavior['Payment Method'].isna().sum()
```

* Column: **Discount Applied**

There are other algorithms for easily and automatically filling missing values, two of them are:  
**backward fill**: fills the missing value with the next non `NaN` value, with `bfill`.  
**forward fill**: fills the missing value with the previous non `NaN` value, with `ffill`.
For this column, we will be using `ffill` passed to the `method` argument:

```python
shopping_behavior['Discount Applied'] = shopping_behavior['Discount Applied'].fillna(method="ffill")
shopping_behavior['Discount Applied'].isna().sum()
```

Great. Now let us check the sum of our missing values again:

```python
total_missing_values = shopping_behavior.isna().sum().sum()
print("total missing values: ", total_missing_values)
```

#### **Wrong Data**

Wrong data are entries with inconsistencies in them. For example, finding entries with negative values for a positive column, wrong dates (eg. dates after today for exents)...
For sure we would want to clean the data from inconsistencies.

```python
shopping_behavior.describe()
```

* As we could notice, **Purchase Amount** has negative values (min is ~(-82))  
Let us fix that by converting all negative values to 0 for this column:

```python
# initialize the condition
conditional_rows = shopping_behavior['Purchase Amount (USD)'] < 0
# assign objects satisfying condition to 0 for that column
shopping_behavior.loc[conditional_rows, 'Purchase Amount (USD)'] = 0
```

```python
# display number of negative values for the column
wrong_amounts = shopping_behavior.loc[conditional_rows, 'Purchase Amount (USD)'].sum()
# we locate all the rows satisfying the condition, and select only the column we are interested in,
# and we count number of rows
wrong_amounts
```

```python
shopping_behavior.Gender.unique()
```

* unique values should only be {Male, Female} for **Gender**. Thus, we should fix that by converting the objects with gender 'female' to 'Female' (uppercase F):

```python
# initialize the condition and apply the changes
conditional_gender = shopping_behavior.Gender == 'female'
shopping_behavior.loc[conditional_gender, 'Gender'] = "Female"

# display the unique values now
shopping_behavior.Gender.unique()
```

#### **Customization**

`pyplot` offers too many ways and methods to customize figures and plots. We will see few of them today (we should always refer to the official documentation to explore more and find what we really are intending to do with a particular plot). 

* **Figure size**, using `figure()` passed to it `figzise` with the needed width and height in inches:

* **Labelling x-axis and y-axis** with `xlabel()` and `ylabel()`:

```python
plt.figure(figsize=[10, 6])
plt.plot(x)
plt.plot(y)

# setting the x-axis label
plt.xlabel('Number of days passed')
# setting the y-axis label
plt.ylabel('Average sales')

plt.show()
```

* **Titling a plot** using `title()`:

```python
plt.figure(figsize=[10, 6])
plt.plot(x)
plt.plot(y)

plt.xlabel('Number of days passed')
plt.ylabel('Average sales')

# setting a title
plt.title('Sales of Product X and Product Y over the days')
plt.show()
```

You can even describe a font style and pass it to the title or label through `fontdict`, and a position for the title with `loc`. All as follows:

```python
# setting font styles
title_style = {'family':'serif','color':'orange','size':18, 'weight':'bold'}
labels_style = {'family':'serif','color':'darkred','size':15}
```

```python
# recreating the plots
plt.figure(figsize=[10, 6])
plt.plot(x)
plt.plot(y)

plt.xlabel('Number of days passed', fontdict=labels_style)
plt.ylabel('Average sales', fontdict=labels_style)

# setting a title
plt.title('Sales of Product X and Product Y over the days', fontdict=title_style, loc='right')
plt.show()
```

* **Gridding the plot**, `grid():`

```python
# recreating the plots
plt.figure(figsize=[10, 6])
plt.plot(x)
plt.plot(y)

plt.xlabel('Number of days passed', fontdict=labels_style)
plt.ylabel('Average sales', fontdict=labels_style)
plt.title('Sales of Product X and Product Y over the days', fontdict=title_style, loc='right')

# adding a grid
plt.grid()

plt.show()
```

**Note** that you can define only x-grid by passing `x` to `axis` (`axis='x'`). Same way with y-grid (`axis='y'`).

* Subplots, inside one Plot, by `subplot()`:

```python
# a subplot for each product
fig, ax = plt.subplots(nrows=1, 
                       ncols=2, 
                       figsize=(16, 8))

ax[0].plot(x)
ax[0].set_title('X')
ax[1].plot(y)
ax[1].set_title('Y')

plt.show()
```

It provides an overview of the distribution's shape and helps identify patterns such as normal distribution, skewed distribution, or multimodal distribution (do not worry, as you move through Studies, you will learn everything).

### Relationships

Relationships plots are visual representations that depict the relationships and associations between two or more variables in a dataset. These plots commonly utilize scatter plots, where each data point is represented by a marker on a Cartesian coordinate system.  
Relationship plots provide insights into the data that can aid in decision-making, hypothesis testing, and feature selection.  
We will talk about two of them today.

#### **Scatter Plot**

Two dimensional plot that displays the relationship between two **continuous variables**. Each data point is represented as a marker on the plot.  
To make a scatter plot, we call the `plot()` function with `kind='scatter'`:

```python
# chose only 100 rows to see the plot clearly
df.iloc[0:100, :].plot(kind='scatter', x='Review Rating', y='Purchase Amount (USD)')
```

As we have no intuitively clearer relationship between our data attributes, let us discover it here:

```python
employee_churn = pd.read_csv('data/employeeChurn.csv', index_col = "Unnamed: 0")
employee_churn.head()
```

We will scatter plot the relationship between **Age** and **MonthlyIncome**

```python
employee_churn.plot(kind='scatter', x='Age', y='MonthlyIncome')
```

Positive Relationship exists there between the two columns.

Now, what if we want to see the gender inside the plot. To see the **Gender** column of the data point in the scatter plot. We do that by passing `c` argument to scatter plot (color of marker or dots). However, this argument takes only numerical values. Thus, we will convert the **binary** attribute Gender to numeric values, 1 for 'Male' and 0 for 'Female'.

```python
employee_churn['Gender_binary'] = [0 if gender == 'Female' else 1 for gender in employee_churn['Gender']]
employee_churn.Gender_binary
```

Great, Now let's pass it to the `c` argument:

```python
employee_churn.plot(kind='scatter', x='Age', y='MonthlyIncome', c='Gender_binary')
```

Ugly colors? let's change them (using `cmap` which maps the numerical values of the attribute passed to `c` to real colors):

```python
employee_churn.plot(kind='scatter', 
                    x='Age', 
                    y='MonthlyIncome', 
                    c='Gender_binary', 
                    cmap='brg')
```

So, the green dots are of Male data objects and blue ones of of Female.

**Note:** 'brg' stands for 'blue-red-green', one of the values `cmap` can take. There are too many. You can explore them on your own.
